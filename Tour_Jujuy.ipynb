```python
import pandas as pd
import numpy as np
import time
import folium
from weather import Weather, Unit
import pyowm
```

```python
tour_data = pd.read_csv('tour_jujuy.csv', delimiter='\t')
tour_data["pronóstico"] = np.nan
owm = pyowm.OWM('c0bd31f75a52fc384cdc874c33dba532')
colores = {'20/01/19': '#f00', '21/01/19':'#ff9a00', '22/01/19': '#fdff00', '23/01/19': '#15ff00'}
```

```python
serie_pronostico = []
for index, row in tour_data.iterrows():
    loc = row['coordenadas'].split(',')
    lat = float(loc[0])
    lng = float(loc[1])
    observation = owm.weather_at_coords(lat, lng)
    pronostico = observation.get_weather().get_temperature('celsius')
    #print(pronostico)
    pronostico_tmp = str(pronostico['temp']) + '°C ' 
    #+ str(pronostico['temp_min']) + '/' + str(pronostico['temp_max']) + '°C'

    print(pronostico_tmp)
    serie_pronostico.append(pronostico_tmp)

    time.sleep(15)
```

```python
tour_data['pronóstico'] = np.array(serie_pronostico)
```

```python
folium_map = folium.Map(location=[-24.1852569, -65.2994789],
                            zoom_start=7)

for index, row in tour_data.iterrows():

    # add marker to the map
    loc = row['coordenadas'].split(',')
    lat = float(loc[0])
    lng = float(loc[1])

    folium.CircleMarker(location=[lat, lng],
                        radius=3,
                        color=colores[row['fecha']],
                        popup=row['hora'] + 
                        '. Pronóstico: ' + row['pronóstico'] +
                        '. Gastos: ' + str(row['gastos']),
                        tooltip=row['descripción'],
                        fill=True).add_to(folium_map)
folium.TileLayer('openstreetmap').add_to(folium_map)

folium_map
```
